```python
def find_var_type(var):
    
    if var in discrete_var:
        print("{} is a numerical variable, Discrete in nature".format(var))
    elif var in cont_var:
        print("{} is a nmerical variable, continuous in nature".format(var))
    else:
        print("{} is a categorical variable".format(var))
```

```python
find_var_type('Cloud3ps')
```

```python
def RandomSampleImputation(df, feature):
    df[feature] = df[feature]
    random_sample = df[feature].dropna().sample(df[feature].isnull().sum(), random_state = 0)
    random_sample.index = df[df[feature].isnull()].index
    df.loc[df[feature].isnull(),feature] = random_sample
```

```python
RandomSampleImputation(df, "Cloud9am")
RandomSampleImputation(df, "Cloud3pm")
RandomSampleImputation(df, "Evaporation")
RandomSampleImputation(df, "Sunshine")
```

```python
df.isnull().sum()*100/len(df)
```

```python
find_var_type('RISK_MM ')
```

```python
def MeanImputation(df, feature):
    df[feature] = df[feature]
    mean = df[feature].mean()
    df[feature] = df[feature].fillna(mean)
```

```python
MeanImputation(df, "Pressure9am")
MeanImputation(df, "Pressure3pm")
MeanImputation(df, "MinTemp")
MeanImputation(df, "MaxTemp")
MeanImputation(df, "Rainfall")
```

## Plotting Q-Q plot

```python
import scipy.stats as stats
import pylab
```

```python
def plot_curve(df, feature):
    plt.figure(figsize = (10, 6))
    plt.subplot(1, 2, 1)
    df[feature].hist()
    
    plt.subplot(1, 2, 2)
    stats.probplot(df[feature], dist = 'norm', plot = pylab)
    plt.title(feature)
    plt.show()
```

```python
for i in cont_var:
    plot_curve(df, i)
```

## Splitting of data into X and y

```python
X = df.drop(['RainTomorrow'], axis = 1)
y = df["RainTomorrow"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
scaled = scaler.transform(X)
```

```python
X.columns
```

```python
X = pd.DataFrame(scaled, columns = X.columns)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
rfc_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
acc_rfc = accuracy_score(y_test, rfc_pred)
print("The accuracy score for RandomForestClassifier is {:.2f}".format(acc_rfc))
```

```python
cm = confusion_matrix(y_test, rfc_pred)
print(cm)
```

```python
cr = classification_report(y_test, rfc_pred)
print(cr)
```

```python
from sklearn import metrics
```

```python
metrics.plot_roc_curve(rfc, X_test, y_test)
```

```python
metrics.roc_auc_score(y_test, rfc_pred, average = None)
```
